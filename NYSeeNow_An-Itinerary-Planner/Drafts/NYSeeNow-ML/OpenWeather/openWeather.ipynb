### How to use:
- Query and store their historical data for 2019 in New York 
- How to query:
- API has max 1,000 calls per day and max of 60 calls per minute
- To store:
- Process similar to last project and store in a POSTGRESQL DB
- There are 8765.999 hours in a year
- Therefore will take 16 minutes to make 960 calls in one day
- It will take 9 days of this get the weather data for every hour in 2019
- Ensure I dont go over the limits

- Create Data Dictionary


```python
import json
import pandas as pd
```

```python
# Extract the information for the current weather and append to a weather data file
with open("weatherData.json") as json_file:
    w_data = json.load(json_file)

#show this data in a data frame
#number of hours in the data
# around 976 per day?
num= 1952
# Create an empty DataFrame to store the venue data
weather_data = pd.DataFrame(columns=['Latitude', 'Longitude', 'Time Stamp','Temperature (C)', 'Weather ID', 'Weather Main', 'Weather Description', 'Weather Icon', 'Rain 1h', 'Snow 1h'])

#select the first venue
for k in range(0,num):
    data=w_data[k]
    lat=data['lat']
    lon=data['lon']
    time_stamp=data['data'][0]['dt']
    temp=data['data'][0]['feels_like']-273.15#store in celsius
    weather_id=data['data'][0]['weather'][0]['id']
    weather_main=data['data'][0]['weather'][0]['main']
    weather_description=data['data'][0]['weather'][0]['description']
    weather_icon=data['data'][0]['weather'][0]['icon']
    try:
        rain_1h=data['data'][0]['rain']['1h']
    except KeyError:
        rain_1h='NA'
    
    try:
        snow_1h=data['data'][0]['snow']['1h']
    except KeyError:
        snow_1h='NA'

    #append the values information to the DataFrame
    weather_data.loc[k] = [lat,lon,time_stamp,temp,weather_id,weather_main,weather_description,weather_icon,rain_1h,snow_1h]


weather_data.head(20)
```
