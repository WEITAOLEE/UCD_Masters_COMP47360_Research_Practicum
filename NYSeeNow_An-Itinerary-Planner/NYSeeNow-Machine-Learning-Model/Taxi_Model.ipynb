```python
from pycaret.datasets import get_data
from pycaret.classification import *
from pycaret.regression import *
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
import glob
li = []
for file in glob.glob('drive/MyDrive/New_Taxi_Data.csv/part*'):
  df = pd.read_csv(file)
  li.append(df)

merged_df = pd.concat(li, axis=0, ignore_index=True)
```

```python
df = merged_df
```

```python
df.drop(["tpep_pickup_datetime", "tpep_dropoff_datetime"], axis=1, inplace=True)
```

```python
new_grouped = df.groupby(['DOLocationID', 'dropoff_day_number', 'dropoff_hour', 'dropoff_month'])['passenger_count'].sum().reset_index()
```

```python
s1 = setup(new_grouped, target='passenger_count',
           fold=5,
           numeric_features=['DOLocationID', 'dropoff_hour', 'dropoff_day_number', 'dropoff_month'],
           session_id=125)
```

```python
best = compare_models()
```

```python
# Tune model
tuned_model = tune_model(best)

# Evaluate model
evaluate_model(tuned_model)

# Finalize model
final_model = finalize_model(tuned_model)
```
